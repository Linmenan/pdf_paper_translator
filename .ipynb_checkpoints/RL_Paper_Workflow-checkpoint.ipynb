```python
import os
import sys
from IPython.display import display, Markdown, Image
from glob import glob
import re

# å‡è®¾æ‚¨å·²å°†æ‰€æœ‰æ ¸å¿ƒåŠŸèƒ½å‡½æ•°æ”¾å…¥ workflow_utils.py æ–‡ä»¶ä¸­
# æ³¨æ„ï¼šæ‚¨éœ€è¦ç¡®ä¿æ­¤æ–‡ä»¶ä¸æ‚¨çš„ Notebook ä½äºåŒä¸€ç›®å½•ä¸‹ï¼Œæˆ–è€…åœ¨ Python è·¯å¾„ä¸­
try:
    import workflow_utils as wf
    print("âœ… æˆåŠŸåŠ è½½ workflow_utils æ¨¡å—ã€‚")
except ImportError:
    display(Markdown("**ğŸ”´ é”™è¯¯ï¼šæ— æ³•å¯¼å…¥ workflow_utils æ¨¡å—ã€‚**"))
    display(Markdown("è¯·ç¡®ä¿æ‚¨å·²å°†æ‰€æœ‰æ ¸å¿ƒå‡½æ•°å®šä¹‰æ•´åˆåˆ°ä¸€ä¸ªåä¸º `workflow_utils.py` çš„æ–‡ä»¶ä¸­ï¼Œå¹¶ç¡®ä¿è¯¥æ–‡ä»¶ä½äºå½“å‰ç›®å½•ã€‚"))
    sys.exit(1)


# --- å…¨å±€é…ç½® ---
# 1. æ–‡çŒ®æ ¹ç›®å½•ï¼šè¯·ç¡®ä¿æ­¤ç›®å½•å­˜åœ¨ï¼Œä¸”åŒ…å«æ‚¨çš„ PDF æ–‡ä»¶
PDF_ROOT_DIR = "./academic_papers" 
os.makedirs(PDF_ROOT_DIR, exist_ok=True)
print(f"æ–‡çŒ®æ ¹ç›®å½•å·²è®¾ç½®ä¸º: {PDF_ROOT_DIR}")

# 2. ç»“æœè¾“å‡ºç›®å½•
OUTPUT_DIR = "./extracted_output"
os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"æå–ä¸ç»“æœè¾“å‡ºç›®å½•: {OUTPUT_DIR}")

# 3. LLM API Key æ£€æŸ¥
if not os.getenv("GEMINI_API_KEY"):
    display(Markdown("**âš ï¸ è­¦å‘Šï¼šGEMINI_API_KEY ç¯å¢ƒå˜é‡æœªè®¾ç½®ï¼è¯·åœ¨å¯åŠ¨ Notebook çš„ç»ˆç«¯ä¸­æ‰§è¡Œ `export GEMINI_API_KEY=\"YOUR_KEY\"` åé‡å¯ Notebook Kernelã€‚**"))
# ----------------------------------
```

```python
# --- Step 1: PDF æ–‡ä»¶é€‰æ‹©ä¸å†…å®¹æå– ---

# 1. æ‰«æç›®å½•ä¸‹çš„æ‰€æœ‰ PDF æ–‡ä»¶
pdf_files = sorted(glob(os.path.join(PDF_ROOT_DIR, "*.pdf")))

if not pdf_files:
    display(Markdown(f"**ğŸ”´ é”™è¯¯ï¼šæœªåœ¨ç›®å½• `{PDF_ROOT_DIR}` ä¸­æ‰¾åˆ°ä»»ä½• PDF æ–‡ä»¶ã€‚**\nè¯·å°†æ‚¨çš„æ–‡çŒ®æ”¾å…¥è¯¥æ–‡ä»¶å¤¹ã€‚"))
    sys.exit(1)

print("è¯·é€‰æ‹©æ‚¨è¦è§£è¯»çš„æ–‡çŒ®ç¼–å·ï¼š")
for i, file_path in enumerate(pdf_files):
    print(f"[{i + 1:2d}] {os.path.basename(file_path)}")
print("-" * 40)

# 2. æ¥æ”¶ç”¨æˆ·è¾“å…¥ï¼ˆä½¿ç”¨ input()ï¼‰
try:
    selection = input("è¯·è¾“å…¥ç¼–å· (è¾“å…¥ q é€€å‡º): ")
    if selection.lower() == 'q':
        print("ç¨‹åºé€€å‡ºã€‚")
        sys.exit(0)
    
    idx = int(selection) - 1
    
    if 0 <= idx < len(pdf_files):
        SELECTED_PDF_PATH = pdf_files[idx]
        SELECTED_FILE_NAME = os.path.basename(SELECTED_PDF_PATH)
        
        # å®šä¹‰æ–‡ä»¶è·¯å¾„å˜é‡ï¼Œç”¨äºä¼ é€’ç»™åç»­ Cell
        LLM_INPUT_PATH = os.path.join(OUTPUT_DIR, SELECTED_FILE_NAME.replace('.pdf', '_llm_input.txt'))
        LLM_RESULT_PATH = LLM_INPUT_PATH.replace('_llm_input.txt', '_result.txt')
        
        # 3. æ‰§è¡Œæå– (è°ƒç”¨ workflow_utils ä¸­çš„å‡½æ•°)
        INPUT_FILE_PATH = wf.extract_text_from_pdf(SELECTED_PDF_PATH, LLM_INPUT_PATH)
        
    else:
        print(f"æ— æ•ˆç¼–å·ã€‚è¯·è¾“å…¥ 1 åˆ° {len(pdf_files)} ä¹‹é—´çš„æ•°å­—ã€‚")
        INPUT_FILE_PATH = None
        
except ValueError:
    print("è¾“å…¥æ— æ•ˆã€‚è¯·è¾“å…¥æ•°å­—æˆ– qã€‚")
    INPUT_FILE_PATH = None

# æ‰“å°è·¯å¾„ä¿¡æ¯ï¼Œç¡®ä¿ä¼ é€’æ­£ç¡®
if INPUT_FILE_PATH:
    print(f"\n[è·¯å¾„å˜é‡è®¾ç½®]\nLLM è¾“å…¥æ–‡ä»¶: {LLM_INPUT_PATH}\nLLM ç»“æœæ–‡ä»¶: {LLM_RESULT_PATH}")
    
# ç¡®ä¿åœ¨åç»­ Cell ä¸­å¯ä»¥è®¿é—®è¿™äº›è·¯å¾„å˜é‡
globals()['INPUT_FILE_PATH'] = INPUT_FILE_PATH
globals()['LLM_RESULT_PATH'] = LLM_RESULT_PATH
```

```python
# --- Step 2: LLM æ¨ç† (Gemini API è°ƒç”¨) ---

if 'INPUT_FILE_PATH' in globals() and INPUT_FILE_PATH:
    # 1. è¯»å–åŒ…å«æŒ‡ä»¤å’Œæ–‡æœ¬çš„å®Œæ•´è¾“å…¥æ–‡ä»¶
    with open(INPUT_FILE_PATH, 'r', encoding='utf-8') as f:
        prompt_content = f.read()
    
    # 2. è¿è¡Œåˆ†æ (è°ƒç”¨ workflow_utils ä¸­çš„å‡½æ•°)
    RESULT_FILE_PATH = wf.run_gemini_analysis(prompt_content, LLM_RESULT_PATH)
    
    # ç¡®ä¿è·¯å¾„è¢«æ­£ç¡®è®¾ç½®å’Œä¼ é€’
    globals()['RESULT_FILE_PATH'] = RESULT_FILE_PATH

else:
    print("ğŸ”´ è·³è¿‡ LLM æ¨ç†ï¼šä¸Šä¸€æ­¥æœªæˆåŠŸç”Ÿæˆè¾“å…¥æ–‡ä»¶ã€‚")
    RESULT_FILE_PATH = None
```

```python
# --- Step 3: å¯è§†åŒ– (Graphviz æ¸²æŸ“) ä¸ç»“æœå±•ç¤º ---

if 'RESULT_FILE_PATH' in globals() and RESULT_FILE_PATH:
    
    # 1. æå– DOT ä»£ç  (è°ƒç”¨ workflow_utils ä¸­çš„å‡½æ•°)
    dot_code = wf.extract_dot_code(RESULT_FILE_PATH)
    
    if dot_code:
        # ä½¿ç”¨ç»“æœæ–‡ä»¶åä½œä¸ºå›¾ç‰‡åç§°çš„åŸºç¡€
        image_name = os.path.basename(RESULT_FILE_PATH).replace('_result.txt', '_structure')
        
        # 2. æ¸²æŸ“å¹¶æ˜¾ç¤ºå›¾ç‰‡ (è°ƒç”¨ workflow_utils ä¸­çš„å‡½æ•°)
        wf.render_and_display_graph(dot_code, image_name, output_dir=OUTPUT_DIR, format='png')
    
    # 3. æœ€åå±•ç¤º LLM çš„åŸå§‹è¾“å‡ºæ–‡æœ¬
    display(Markdown("--- \n## ğŸ“œ LLM åŸå§‹æ–‡æœ¬è¾“å‡º (ç»“æ„æ€»ç»“ä¸ç¿»è¯‘) \n---"))
    try:
        with open(RESULT_FILE_PATH, 'r', encoding='utf-8') as f:
            full_result = f.read()
            display(Markdown(full_result))
    except Exception as e:
        print(f"ğŸ”´ æ— æ³•è¯»å–ç»“æœæ–‡ä»¶è¿›è¡Œæ˜¾ç¤ºï¼š{e}")

else:
    print("ğŸ”´ è·³è¿‡å¯è§†åŒ–ï¼šä¸Šä¸€æ­¥æœªæˆåŠŸç”Ÿæˆ LLM ç»“æœæ–‡ä»¶ã€‚")
```
